```python
import geopandas as gpd
import pandas as pd
```

```python
url2='https://raw.githubusercontent.com/bbh-llc/maps/master/ess_payout_locations/data/DA-20-823A2.geojson'
df2 = gpd.read_file(url2)
```

```python
df2.columns
```

```python
df2_dissolved = df2.dissolve(by='ApplicantName', aggfunc='count')
```

```python
len(df2_dissolved)
```

```python
df2_dissolved.sort_values(by='Callsign', ascending=False, inplace=True)
```

```python
df2_dissolved.head(20)
```

```python
df2_dissolved.index[0]
```

```python
df2_dissolved['Callsign'].sum()
```

```python
3401/16458
```

```python
key_words = ['Christian', 'Bible', 'Bishop', 'Church', 'Faith', 'Evangelical', 'Christ']
```

```python
def religious_entity(x):
    words = x.split()
    for word in words:
        if word not in key_words:
            continue
        else: 
            return True
```

```python
result=[religious_entity(x) for x in df2_dissolved.index]
```

```python
evangelical = pd.Series(result)
```

```python
evangelical
```

```python
len(df2_dissolved)
```

```python
df2_dissolved['Evangelical'] = [x for x in evangelical]
```
